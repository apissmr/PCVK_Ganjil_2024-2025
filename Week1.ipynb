```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# membuat list untuk menyimpan url dari beberapa image

urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://i.pinimg.com/originals/23/12/4f/23124fe49e451c20c99c4817043db06a.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  try: # this is a try-except block to catch the error
    image   = io.imread(url)
    image   = cv.resize(image, (0,0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((image, image_2))
    cv2_imshow(final_frame)
    print('\n')
  except Exception as e: # this code will run if the code in the try block throws an exception
    print(f"Error loading image from {url}: {e}") # prints the url and the message
```

```python
#melihat ukuran File Image
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [135,99,59]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#TUGAS PRAKTIKUM 1
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertikal ditengah image
for y in range (tinggi):
  image_3[y, int((lebar)/2)] = [200,99,59]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Calculate the slope of the diagonal line
slope = tinggi / lebar

# Draw the diagonal cross lines
for x in range(tinggi):
  for y in range(lebar):
    if abs(x - slope * y) < 1 or abs(x - (tinggi - slope * y)) < 1 :
      image_3[x, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
